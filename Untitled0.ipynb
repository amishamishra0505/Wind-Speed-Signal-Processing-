```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import pywt
```

```python
data_set=pd.read_csv("monthlydataset.csv")
```

```python
data_set
```

```python
data=data_set[["Source","Location ID","City", "State", "Relative Humidity Units"]]
```

```python
data
```

```python
data= data.drop(labels=[0])
```

```python
data
```

```python
data=data.rename(columns = {"Source": "Year", "Location ID": "Month",
                              "City": "Day", 
                              "State": "Hour",
                              "Relative Humidity Units": "Wind Speed"})
```

```python
data
```

```python
data
```

```python
data['dateInt']=data['Year'].astype(str).str.zfill(2) + data['Month'].astype(str).str.zfill(2) + data['Day'].astype(str).str.zfill(2)+ data['Hour'].astype(str).str.zfill(2)
data['Date'] = pd.to_datetime(data['dateInt'], format='%Y%m%d%H')
```

```python
data
```

```python
data= data.drop(labels=["dateInt"], axis=1)
```

```python
data
```

```python
data=data[["Year","Month","Day","Hour","Date","Wind Speed"]]
```

```python
data
```

```python
df=data[['Date','Wind Speed']]
df
```

```python
df=df.set_index('Date')
df
```

```python
df.dtypes
```

```python
df = df.astype({"Wind Speed": float})
df
```

```python
df['Wind Speed'].plot();
```

```python
ax = df.loc['2000-01':'2000-02', 'Wind Speed'].plot(linestyle='-')
ax.set_ylabel('Wind Speed (m/s)');
```
